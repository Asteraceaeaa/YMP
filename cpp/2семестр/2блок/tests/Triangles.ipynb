```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

def displayTriangles(verticesABC, verticesP1P2P3):

    polyABC = Polygon(verticesABC, edgecolor="gray")
    polyP1P2P3 = Polygon(verticesP1P2P3, edgecolor="black", facecolor="green", zorder=2)

    fig, ax = plt.subplots()
    ax.add_patch(polyABC)
    ax.add_patch(polyP1P2P3)

    for coord, label in zip(verticesABC + verticesP1P2P3, ["A", "B", "C"] + ["P1", "P2", "P3"]):
        x, y = coord
        plt.plot(x, y, 'o', label=f"{label}({x}, {y})")

    ax.set_aspect("equal")

    xmax = max([p[0] for p in verticesP1P2P3 + verticesABC])
    ymax = max([p[1] for p in verticesP1P2P3 + verticesABC])
    ax.set_xlim(0, xmax + 1.2)
    ax.set_ylim(0, ymax + ymax / 5)
    ax.legend(fontsize=12)
    plt.show()
    
```

```python


# Вершины для true
verticesABC = [(0.2, 0.2), (0.8, 0.2), (0.5, 0.8)]
verticesP1P2P3 = [(0.3, 0.25), (0.5, 0.3), (0.5, 0.6)]

displayTriangles(verticesABC, verticesP1P2P3)
```

```python

# Вершины для false
verticesABC = [(0.2, 0.2), (0.8, 0.2), (0.5, 0.8)]
verticesP1P2P3 = [(0.3, 0.25), (0.9, 0.5), (0.5, 0.6)]

displayTriangles(verticesABC, verticesP1P2P3)
```
