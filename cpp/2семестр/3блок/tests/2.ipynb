```python
import numpy as np

def distance_to_line(point, line_start, line_end):
    """
    Рассчитывает расстояние от точки до БЕСКОНЕЧНОЙ прямой.
    Работает на основе векторного произведения.
    """
    p = np.array(point)
    a = np.array(line_start)
    b = np.array(line_end)

    # Векторное произведение для нахождения площади параллелограмма
    # Норма вектора (b-a) в знаменателе
    distance = np.abs(np.cross(b-a, p-a)) / np.linalg.norm(b-a)
    return distance

# Использование
point = (1, 4)
line_start = (1, 1)
line_end = (2, 2)

dist = distance_to_line(point, line_start, line_end)
print(f"Расстояние до бесконечной прямой: {dist}") # Вывод: 0.0
```
