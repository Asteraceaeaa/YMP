```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
fig, axes = plt.subplots(ncols=2, nrows=1)

#1

P = (-1.53, 3.61)

circle = Circle((1.3, 0.78), 4, fill=False)
axes[0].add_patch(circle)

axes[0].scatter(*P, s=2, color="Red")
#2
O1 = (4, 3.2); R1 = 5
O2 = (2, 4); R2 = 2

c1 = Circle(O1, R1, fill=False)
c2 = Circle(O2, R2, fill=False)

axes[1].add_patch(c1)
axes[1].add_patch(c2)

Ox = [x[0] for x in [O1, O2]]
Oy = [x[1] for x in [O1, O2]]

axes[1].scatter(Ox, Oy)

for ax in axes:
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.set_aspect("equal")
plt.show()
```
